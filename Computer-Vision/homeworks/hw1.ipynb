## 视觉作业1：

对于输入的一幅图像编程实现以下功能:

1）计算图像的直方图

2）进行直方图均衡处理，输出处理后的直方图

3）实现一种图像滤波，并输出滤波后的图像及其直方图

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./img/Lenna.png', cv2.IMREAD_GRAYSCALE)
```

```python
# 1）计算图像的直方图
histogram, bins = np.histogram(image.flatten(), 256, [0, 256])
# 定义累积分布函数 cdf（Cumulative Distribution Function)
cdf = histogram.cumsum()
plt.figure()
plt.title("Original Histogram")
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
# 2）进行直方图均衡处理
equ_image = cv2.equalizeHist(image)

# 输出处理后的直方图
plt.figure()
plt.title("Equalized Histogram")
plt.hist(equ_image.ravel(), 256, [0, 256])
plt.show()
```

```python
# 3）实现一种图像滤波（这里使用的是 Gaussian Blur）
filtered_image = cv2.GaussianBlur(image, (5, 5), 0)

plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image')
plt.show()
```

```python
plt.figure()
plt.subplot(1, 2, 1) 
plt.title("Original Histogram")
plt.hist(image.ravel(), 256, [0, 256])

# 滤波后的直方图
plt.subplot(1, 2, 2)
plt.title("Filtered Image Histogram")
plt.hist(filtered_image.ravel(), 256, [0, 256])

plt.show()
```

滤波后的直方图变化：

- **灰度级平滑**：高斯滤波使得图像的灰度级别更加平滑。
- **噪声减少**：滤波后，噪声（直方图的尖峰或者波动）减少，直方图会变得更加平滑。
- **对比度变化**：由于高斯滤波是一种平滑技术，它可能会降低图像的局部对比度。对比滤波前后的灰度直方图，可以发现滤波后灰度级别的分布变得更加集中，而不是广泛分布。
